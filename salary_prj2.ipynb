```python
## Now  Model Building
def model_building():
    df.shape
    X= df['YearsExperience']   
    y=df['Salary']
    X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8)
    X_train.shape,y_train.shape
    X_test.shape,y_test.shape
    X_train = X_train.values.reshape(-1,1)
    y_train = y_train.values.reshape(-1,1)
    
    model = LinearRegression()
    model_fit = model.fit(X_train,y_train)
    coef= model_fit.coef_
    intercept=model_fit.intercept_
    
    X_test = X_test.values.reshape(-1,1)
    
    y_pred = model_fit.predict(X_test)
    print(np.round(y_pred))
   
    #Model evaluation
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print("                     ")
    print("Coeff:",coef)
    print("Interception:", intercept)
    print(f'Mean Squared Error: {mse:.2f}')
    print(f'R-squared: {r2:.2f}')
```

```python
dmb= model_building()
print(dmb)
```

```python
print(f'Coefficient:')
print("\nCoefficient: The coefficient of the model represents the estimated change in salary for a one-unit increase in years of experience. In this case, it indicates the average increase in salary for each additional year of experience.")
```

```python

print(f'Intercept:')
print("\nThe intercept represents the estimated starting salary when years of experience are zero. It contributes to the linear regression model.")
```

```python
print(f'Mean Squared Error: ')
print("The mean squared error provides a measure of the average squared difference between actual and predicted salaries. A lower MSE indicates better model performance.")
```

```python
print(f'R-squared: ')
print("\nR-squared measures the proportion of the variance in the dependent variable (Salary) that is predictable from the independent variable (YearsExperience).")
print("It implies a strong correlation between the predictor and the response variable, indicating that the model does an excellent job of explaining the variability in the data")
```
